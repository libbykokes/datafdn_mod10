## Question 1: Read the crunchbase_odm_orgs.csv CSV file into a Pandas data frame and drop records where company names are None values:
- Print the count of records (rows) in the resulting data frame
- Print the first 5 records of the data frame


```python
import pandas as pd

df = pd.DataFrame(pd.read_csv("crunchbase_odm_orgs.csv"))
print("The number of null records is", df["name"].isna().sum())
print("Meanwhile the number of records whose company name is actually 'None' is", (df["name"] == "None").sum())
nonnull = df.dropna(subset=["name"])
print("The CSV contains", len(nonnull), "rows") # Count records
nonnull.head()
```

## Question 2: Create a data frame that contains only the records of USA-based companies whose name starts with "Ac":
- Print the count of records (rows) in the resulting data frame
- Print the first 5 records of the data frame


```python
usaac = nonnull[(nonnull["country_code"] == "USA") & (nonnull["name"].str.contains(r"^Ac", regex=True))] # carrot means 'starts with'
print("The resulting rows of the subset dataframe are", len(usaac))
usaac.head()
```

## Question 3: Convert the data frame from the previous step into a list of JSON objects:
- Print the count of JSON objects
- Print the first 5 JSON objects
- Write all JSON objects into a text (JSON string) file
- Print the number of records in the resulting file


```python
import json

usaac_json = json.loads(usaac.to_json(orient="records")) # Given that orientation, each list element is 1 company
print("The count of JSON objects is", len(usaac_json))
print("\nAnd the original dataframe record count was", len(usaac)) # Get confirmation

print("\nHere are the first 5 records:\n", json.dumps(usaac_json[:5], indent= 2)) 
# Slicing first 5 records and making them readable

with open("US_Based_Ac_Companies.json", "w") as f:
    json.dump(usaac_json, f, indent=2)
    
print("\nThe record count of this newly-saved file hasn't changed, it's still", len(usaac_json), "rows long")
    
```

## Question 5: Implement Flask API functions to Return a list of Crunchbase companies based in a user submitted City from the original dataset (4 Points)
- Take a user input through a form to return a list of JSON objects from the city queried

I could just create a text box for the user's input, but they're doubtfully going to enter a city value that actually matches the format of the cities that are values in the dataset. So let's just give them a drop-down reference of which city they want to look at companies from. 

```python
# Going to use same server/port as previous cell and then I can visit both URls after I run this cell. 
@app.route('/cityofchoice', methods=['GET', 'POST']) # This route accepts GET for visiting page, POST for submitting form
def form():
    cities = sorted(nonnull["city"].dropna().unique().tolist()) # Alphabetize list of unique non-null city values
    if request.method == 'GET': # Checks how page is accessed
        options = "".join([f'<option value="{city}">{city}</option>' for city in cities])
        # use f-string with a list comprehension dynamically builds an HTML option tag for every city in the list
        # the value= gets sent to Flask on submission, the text between tags shows to the form submitter
        # "".join() joins all the tags in one big spaceless HTML block
        return f'''
            <form method="POST">
                <label for="city">Select a city:</label>
                <select name="city" id="city">
                    {options}
                </select>
                <button type="submit">Submit</button>
            </form>
        ''' # on the 'select' line, name="city" is the key we use to get values, 
            # Select renders dropdown menu instead of a free-form text box 
    else:
        try:
            city = request.form['city']
            city_df = nonnull[nonnull["city"] == city] # Filters OG dataframe for whatever the user input
            city_json = json.loads(city_df.to_json(orient="records")) # Converts the df output to JSON objects, one per company
            return json.dumps(city_json, indent=2) # Sends formatted JSON back to browser as a response
        except KeyError: # If the city key is for some reason missing from form data, print the below message
            return "Please enter a city in the form."

if __name__ == "__main__":
    app.run(host='localhost', port=1997) # Using this server for both Question 5 and 6
    # So then I visited http://localhost:1997/nyc AND http://localhost:1997/cityofchoice
    # Man there are a lot of cities to choose from...
```
